**PHASE 6 — FEATURE ENGINEERING**

Step 18: Attention Score Function

```python
def calculate_attention_score(row):
    score = 0
    score += row['focus_duration'] * 10
    score += row['watch_long_video'] * 10
    score += row['long_content_frequency'] * 8
    score -= row['short_content_time'] * 7
    score -= row['skip_long_videos'] * 8
    score -= row['task_switching'] * 7
    score -= row['notification_interrupt'] * 6
    return max(0, min(100, score + 50))
```

Step 19: Apply Score

```python
# Create a mapping of conceptual names to actual encoded column names
column_mapping = {
    'focus_duration': ' Q7. How long can you stay focused on one task without checking your phone?  ',
    'watch_long_video': 'Q11. Can you comfortably watch a 20–30 minute educational video without losing interest?  ',
    'long_content_frequency': 'Q5. How often do you watch long-form content such as lectures, podcasts, documentaries, or full YouTube videos?  ',
    'short_content_time': 'Q4. How much time do you spend daily watching short-form content (Reels, Shorts, etc.)?  ',
    'skip_long_videos': 'Q8. How often do you skip or fast-forward videos longer than 5 minutes?  ',
    'task_switching': 'Q10. How often do you switch between apps/tasks while doing something important?  ',
    'notification_interrupt': 'Q9. How often do notifications interrupt your concentration while studying, working, or watching content?  '
}

# Invert the mapping for renaming: {actual_column_name: conceptual_name}
inverted_column_mapping = {v: k for k, v in column_mapping.items()}

# Create a temporary DataFrame with renamed columns for the apply function
temp_df = df.rename(columns=inverted_column_mapping)

# Apply the function to the temporary DataFrame
df['attention_score'] = temp_df.apply(calculate_attention_score, axis=1)
```

Step 20: Visualize Score

```python
plt.hist(df['attention_score'])
plt.show()
```

Step 21: Add Category

```python
def categorize(score):
    if score < 40: return 0
    elif score < 70: return 1
    else: return 2

df['attention_category'] = df['attention_score'].apply(categorize)
```

**PHASE 7 — MODEL BUILDING**

Step 22: Random Forest





```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

Step 23: Predictions

```python
y_pred = model.predict(X_test)
```

Step 24: Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

Step 25: Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Step 26: Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
plt.show()
```

**PHASE 9 — FINAL SYSTEM**

Step 30: Recommendation Engine

```python
def recommend(score):
    if score < 40:
        return "Reduce distractions"
    elif score < 70:
        return "Maintain balance"
    else:
        return "Focus on deep work"
```

Step 31: Final Prediction System

```python
def full_system(sample):
    pred = model.predict([sample])[0]
    # Create a Series with the original column names from X
    sample_series = pd.Series(sample, index=X.columns)
    # Rename the relevant columns in the sample_series to their conceptual names
    renamed_sample_series = sample_series.rename(index=inverted_column_mapping)
    score = calculate_attention_score(renamed_sample_series)
    rec = recommend(score)
    return pred, score, rec
```

Step 32: Test Output


```python
sample = X.iloc[0].values
full_system(sample)
```
